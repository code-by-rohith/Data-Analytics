```python
import sympy as sp
from sympy import * 

```

```python

a, b, c = sp.symbols('a b c')
x = sp.symbols('x')
```

```python
x1, y1 = 1, 2
x2, y2 = 2, 3
x3, y3 = 3, 5
```

```python
equation1 = sp.Eq(y1, a*x1**2 + b*x1 + c)
equation2 = sp.Eq(y2, a*x2**2 + b*x2 + c)
equation3 = sp.Eq(y3, a*x3**2 + b*x3 + c)
equation1=sp.symbols("x")
```

```python
solution = sp.solve([equation1, equation2, equation3], (a, b, c))

quadratic_equation = a*x**2 + b*x + c
quadratic_equation = quadratic_equation.subs(solution)

quadratic_equation
```
